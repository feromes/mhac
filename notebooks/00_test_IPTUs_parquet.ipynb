```python
total = len(df)

n_nulos = df["fracao_ideal"].isna().sum()
n_zero = (df["fracao_ideal"] == 0).sum()
n_maior_1 = (df["fracao_ideal"] > 1).sum()
n_valida = ((df["fracao_ideal"] > 0) & (df["fracao_ideal"] <= 1)).sum()

pd.DataFrame({
    "casos": [
        "total_registros",
        "fracao_ideal_nula",
        "fracao_ideal_zero",
        "fracao_ideal_maior_que_1",
        "fracao_ideal_valida_(0,1]"
    ],
    "quantidade": [
        total,
        n_nulos,
        n_zero,
        n_maior_1,
        n_valida
    ],
    "percentual_%": [
        100,
        n_nulos / total * 100,
        n_zero / total * 100,
        n_maior_1 / total * 100,
        n_valida / total * 100
    ]
})
```

```python
AREA_COLS = [
    "area_do_terreno",
    "area_construida",
    "area_ocupada",
]
```

```python
total = len(df)

rows = []

for col in AREA_COLS:
    n_nulo = df[col].isna().sum()
    n_zero = (df[col] == 0).sum()
    n_nulo_ou_zero = ((df[col].isna()) | (df[col] == 0)).sum()
    n_valido = ((df[col] > 0)).sum()

    rows.append({
        "atributo": col,
        "total_registros": total,
        "nulos": n_nulo,
        "zeros": n_zero,
        "nulos_ou_zero": n_nulo_ou_zero,
        "validos_>0": n_valido,
        "nulos_%": n_nulo / total * 100,
        "zeros_%": n_zero / total * 100,
        "nulos_ou_zero_%": n_nulo_ou_zero / total * 100,
        "validos_%": n_valido / total * 100,
    })

summary_df = pd.DataFrame(rows)
summary_df
```
